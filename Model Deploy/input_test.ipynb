```python
df_test = pd.read_csv('../datasets/hospital_augment_test6_3.csv')
```

```python
text_int = list(df_test['gen_data'])
```

```python
test_y = list(df_test['true_data'])
```

```python
pred_test =[]
start_time = time.time()
for i,name in enumerate(text_int):
    print(name, '>', auto_correct_text.generate_correct_text(name),'>',test_y[i])
    pred_test.append(auto_correct_text.generate_correct_text(name))
print("--- %s seconds ---" % (time.time() - start_time))    
```

```python
df_test = pd.DataFrame(pred_test)
```

```python
df_test.columns = ['pred_test']
```

```python
df_test.to_csv('Hospital_Predict11_with_auto_correct.csv',index=False)
```

```python
count =0
wrong_pred =[]
wrong_true =[]
wrong_input = []
for i,key in enumerate(test_y):
    if key.replace('\u200b', '') == pred_test[i].replace('\u200b', '') :
        count += 1
    else:
        wrong_pred.append(pred_test[i].replace('\u200b', ''))
        wrong_true.append(key)
        wrong_input.append(text_int[i])
```

```python
#### Incorrect sentence ####
for i,value in enumerate(wrong_input):
    print('Input:', wrong_input[i])
    print('Prediction:', wrong_pred[i])
    print('Ground Thrust:', wrong_true[i])
    print('-')
```
